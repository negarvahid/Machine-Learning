```python
import glob
from PIL import Image

data_dir = 'chest_xray'
paths = glob.glob(data_dir + '/*/*.jpeg')
# reproducible on different os
paths.sort()
Image.open(paths[0])
```

```python
from sklearn.model_selection import train_test_split
train_paths, test_paths = train_test_split(paths, test_size=0.20, random_state=1337)
train_paths, val_paths = train_test_split(train_paths, test_size=0.20, random_state=1337)
print(f'total size: {len(paths)}, train size: {len(train_paths)}, val size: {len(val_paths)}, test size: {len(test_paths)}')
```

```python
import os 

def create_image_folder(name, paths):
    !mkdir -p {name}
    !mkdir {name}/normal
    !mkdir {name}/pneumonia
    for path in paths:
        dirs = path.split('/')
        dirs[0] = name
        new_path = '/'.join(dirs)
        !cp {path} {new_path}
    return

train_dir = 'xray/train'
val_dir = 'xray/val'
test_dir = 'xray/test'

create_image_folder('xray/train', train_paths)
create_image_folder('xray/val', val_paths)
create_image_folder('xray/test', test_paths)
```

```python
from torchvision import transforms

train_preprocess = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
```

```python
img_dataset = datasets.ImageFolder(train_dir, train_preprocess)
img_dataset
```

```python
dataloader = torch.utils.data.DataLoader(img_dataset, batch_size=4, shuffle=True, num_workers=4)
```
